```python
# 0. 사용할 패키지 불러오기
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
import pandas as pd
import random

# 손실 이력 클래스 정의
class LossHistory(keras.callbacks.Callback):
    def init(self):
        self.losses = []
        
    def on_epoch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        
MAX_IDX = 275
MAX_GRID = 275
FEATURE_NUM = 50
TOTAL_DATA = MAX_GRID * MAX_IDX

# 랜덤시드 고정시키기
np.random.seed(5)

# 1. 데이터 준비하기
dataset = np.loadtxt("./data/made_wifi1.csv", delimiter=",")
# dataset = pd.read_csv("./data/made_wifi4.csv", delimiter=",")
# dataset = dataset.values

# 2. 데이터셋 생성하기
x_train = dataset[:60500,0:FEATURE_NUM]
y_train = []
for x in range(60500):
    y_train.append(dataset[x][FEATURE_NUM] - 1)
y_train = to_categorical(y_train, num_classes=275)
x_test = dataset[60500:,0:FEATURE_NUM]
y_test = []
for x in range(60500,75625):
    y_test.append(dataset[x][FEATURE_NUM] - 1)    
y_test = to_categorical(y_test, num_classes=275)

## 3. 모델 구성하기 // 새로운 모델 구성
model = Sequential()
model.add(Dense(10, kernel_regularizer=keras.regularizers.l2(0.001),
                input_dim=FEATURE_NUM, activation='relu')) # 원래 12
model.add(Dropout(0.3))
# model.add(Dense(55, kernel_regularizer=keras.regularizers.l2(0.001),
#                 activation='relu')) # 원래 12
# model.add(Dropout(0.3))
model.add(Dense(128, kernel_regularizer=keras.regularizers.l2(0.001),
                activation='relu')) # 원래 12
model.add(Dropout(0.3))
# model.add(Dense(8, activation='relu')) # 원래 8
model.add(Dense(275, activation='softmax'))

# 4. 모델 학습과정 설정하기
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# #저장된 모델 학습할 때 
# from keras.models import load_model
# model = load_model('wifi_reg_perceptron_model6.h5')

history = LossHistory() # 손실 이력 객체 생성
history.init()

# 5. 모델 학습시키기
model.fit(x_train, y_train, epochs=2000, batch_size=275, callbacks=[history]) # 원래 1500, 64

# 6. 학습과정 살펴보기
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(history.losses)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

# 7. 모델 평가하기
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))

from keras.models import load_model
model.save('wifi_reg_perceptron_model1.h5')
```
